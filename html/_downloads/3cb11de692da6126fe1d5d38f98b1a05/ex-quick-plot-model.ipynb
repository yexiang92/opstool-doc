# Quick Model and Eigen Visualization

At any point during model creation, you can run the
`opstool.vis.plotly.plot_model`{.interpreted-text role="func"} or
`opstool.vis.pyvista.plot_model`{.interpreted-text role="func"} to
visualize the current model\'s geometric details.

Although both are based on different engines, `PyVista` and `Plotly`,
they provide almost identical APIs. Therefore, you can choose either for
visualization.


# Plotly-based Visualization


```python
import opstool as opst
import opstool.vis.plotly as opsvis

opst.load_ops_examples("ArchBridge2")  # Built-in example model in opstool
# or your model code here
```

Here, we use a built-in example from `opstool`, which is an example of a
deck arch bridge model primarily composed of frame elements and shell
elements. For your own model script, simply place the
`opstool.vis.plotly.plot_model`{.interpreted-text role="func"} function
anywhere you need visualization.


We use the `opstool.vis.plotly.set_plot_props`{.interpreted-text
role="func"} function to predefine some common visualization properties,
which will affect all subsequent visualizations of models, eigenvalues,
and responses.


```python
opsvis.set_plot_props(point_size=0, line_width=3)
```

# Model Geometry Visualization

Visualization is straightforward; simply call the
`opstool.vis.plotly.plot_model`{.interpreted-text role="func"} function.
For details on various parameters, refer to the
`opstool.vis.plotly.plot_model`{.interpreted-text role="func"}.


```python
fig = opsvis.plot_model(show_outline=True)
fig
# fig.show(renderer="browser")  # for interactive use
```

The function `opstool.vis.plotly.plot_model`{.interpreted-text
role="func"} returns a [Plotly
Figure](https://plotly.com/python-api-reference/generated/plotly.graph_objects.Figure.html)
object, which you can visualize using the [show]{.title-ref} method. In
this example, the rendering mode is set to [notebook]{.title-ref},
allowing the figure to be embedded directly within a Jupyter Notebook.
If you want it to automatically choose the rendering mode, simply call
[show]{.title-ref} without providing any arguments. For more details on
the available [renderer]{.title-ref}, please refer to [Displaying
Figures in Python](https://plotly.com/python/renderers/)
